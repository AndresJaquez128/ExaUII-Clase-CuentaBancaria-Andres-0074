#### Crear la clase "CuentaBancaria", con 3 atributos (titular, numeroCuenta, saldo), 2 funciones leerDatos(), mostrarDatos() y 2 objetos (cuenta1, cuenta2).
#### no usar while true, if for listas y diccionarios

```python
# =========================================================
# CLASE: CuentaBancaria
# ATRIBUTOS: titular, numeroCuenta, saldo
# DESARROLLADOR: Jaquez Andres NL = 27
# FECHA: 30 / Abril / 2026
# =========================================================

class CuentaBancaria:
    def __init__(self, titular, numeroCuenta, saldo):
        self.titular = titular
        self.numeroCuenta = numeroCuenta
        self.saldo = saldo

    def leerDatos(self):
        # El programa se detiene para pedir datos por teclado
        print("--- CAPTURA DE DATOS BANCARIOS ---")
        self.titular = input("Nombre del titular: ")
        self.numeroCuenta = input("Número de cuenta: ")
        self.saldo = input("Saldo inicial: ")

    def mostrarDatos(self):
        # Imprime la información almacenada en el objeto
        print("\n--- ESTADO DE LA CUENTA ---")
        print("TITULAR DE LA CUENTA: ", self.titular)
        print("NÚMERO DE CUENTA:     ", self.numeroCuenta)
        print("SALDO ACTUAL:         $", self.saldo)

# --- FIN DE LA CLASE ---

# Creando los 2 objetos (cuenta1, cuenta2) con valores iniciales
cuenta1 = CuentaBancaria("", "", 0)
cuenta2 = CuentaBancaria("", "", 0)

# Interacción para la cuenta 1
print("DATOS CUENTA #1")
cuenta1.leerDatos()
cuenta1.mostrarDatos()
print(60 * "=")

# Interacción para la cuenta 2
print("DATOS CUENTA #2")
cuenta2.leerDatos()
cuenta2.mostrarDatos()
print(60 * "=")

# Pie de software
print("Software desarrollado por Jaquez Andres NL = 27")
```
